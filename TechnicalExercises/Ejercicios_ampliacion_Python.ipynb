## Ejercicio 1.1

* Genera una función que nos diga si un número es múltiplo de otro. 
* Ponla a prueba pidiendo por pantalla al usuario que introduzca los dos números al llamar al método.
* Prepara la petición por pantalla para que si el usuario no introduce números enteros y distintos del 0 le pida que vuelva a introducirlo, poniéndolo nuevamente a prueba.

```python
def is_multiple(): 
    number_one=input('Introduce number one:')
    number_two=input('Introduce number two:')
    while (number_one.isdigit()==False) or (number_one=='0') or (number_two.isdigit()==False) or (number_two=='0'):
        number_one=input('Introduce number one:')
        number_two=input('Introduce number two:')
    else:
        number_one=int(number_one)
        number_two=int(number_two)
        if number_one%number_two==0:
            print(f'{number_one} es múltiplo de {number_two}')
        else:
            print(f'{number_one} NO es múltiplo de {number_two}')
```

## Ejercicio 1.3

El siguiente ejercicio no es mío, lo hice hace muchos años, es muy típico la verdad y principalmente es útil. Hay muchas versiones de él en internet, es largo así que no os preocupéis, os recomiendo dejarlo para los últimos.

Vamos a trabajar ahora creando un montón de funciones para poder trabajar con fracciones de la forma a/b. Esto yo lo he usado para crear clases, pero me parece más útil que lo usemos como varios ejercicios de funciones seguidos, que tampoco tenéis tanto tiempo. Para representar una fracción vamos a utilizar dos enteros: numerador (la 'a') y un denominador (la 'b').  

Así pues, nuestro trabajo se basa en conseguir generar las siguientes funciones (la idea es que cada función nos devuelva un par de valores: el numerador y el denominador):

* leer_fraccion: Queremos que el usuario introduzca 'a' y 'b' por pantalla. Además, queremos que simplifiques la fracción. Para ello tendrás que crear otra función llamada simplificar_funcion, que haréis en el siguiente punto. 
* simplificar_fracción: Esto se hace dividiendo 'a' y 'b' entre el máximo común divisor (MCD) de ellos dos. Os daré esa función para ir más rápido. La idea es que esta función nos devuelva los dos valores ya arreglados. IMPORTANTE: ESTA SERÁ LA FUNCIÓN QUE LLAMAMEMOS CUANDO QUERAMOS SIMPLIFICAR OTRA.
* escribir_fraccion: Ahora queremos simplemente que nos imprima la fracción de la forma 'a/b', pero como sabéis, si el denominador es 1, entonces con que se imprima 'a' es suficiente. 
* sumar_fracciones: Dadas dos fracciones 'a/b' y 'c/d', queremos sumarlas. Para ello hay que recurrir a las fórmulas que vimos hace años en secundaria, que tenéis por Google o que os dejo aquí delante: numerador = a·d+b·c, denominador = b·d. Por supuesto debemos simplificar nuevamente el resultado antes de devolverlo.
* restar_fracciones: Literalmente lo que acabáis de hacer pero cambiando el signo '+' por el signo '-'. 
* multiplicar_fracciones: Aquí nos toca multiplicar numeradores por un lado, denominadores por otro y luego simplificar el resultado.
* dividir_fracciones: Dividir entre una fracción es multiplicar por la inversa. Así, el numerador tendrá que ser a·d y el denominador será b·c. Por supuesto toca simplificar el resultado.   

Finalmente comprobad que funcionan estas funciones. 

* Clase Fracción

```python
class fraccion:
    def __init__(self):
        pass

    def simplify_frac(a,b):
        """
        Toma el numerador y denominador de una fracción y lo simplifica por mínimo común múltiplo.
        """
        from math import gcd as MCD
        c=int(a/MCD(a,b))
        d=int(b/MCD(a,b))
        return c,d

    def escribir_frac(a,b):
        """
        Argumentos:
        * a: numerador
        * b: denominador
        Te devuelve una fracción.
        """
        if b==1:
            print(f'El resultado es {a}')
        else:
            print(f'El resultado es {a}/{b}')
    
    def leer_fraccion():
        """
        Te permite convertir dos números en numerador y denominador para leerlos en forma de fracción.
        """
        a=input('Input numerator:')
        b=input('Input denominator:')
        while (a.isdigit()==False) or (a=='0') or (b.isdigit()==False) or (b=='0'):
            a=input('Input numerator:')
            b=input('Input denominator:')
        else:
            a=int(a)
            b=int(b)
            x,y=fraccion.simplify_frac(a,b)
            return fraccion.escribir_frac(x,y)
       
    def sumar_frac():
        """
        Argumentos:
        * a: numerador fracción 1
        * b: denominador fracción 1
        * c: numerador fracción 2
        * d: denominador fracción 2
        Te suma las dos fracciones. Te devuelve una nueva fracción simplificada por mínimo común múltiplo.
        """
        a=input('Input numerator 1:')
        b=input('Input denominator 1:')
        c=input('Input numerator 2:')
        d=input('Input denominator 2:')
        while (a.isdigit()==False) or (a=='0') or (b.isdigit()==False) or (b=='0') or (c.isdigit()==False) or (c=='0') or (d.isdigit()==False) or (d=='0'):
            a=input('Input numerator 1:')
            b=input('Input denominator 1:')
            c=input('Input numerator 2:')
            d=input('Input denominator 2:')
        else:
            a=int(a)
            b=int(b)
            c=int(c)
            d=int(d)

            x=int((a*d)+(c*b))
            y=int(b*d)
            x,y=fraccion.simplify_frac(x,y)
            return fraccion.escribir_frac(x,y)
        
    def restar_frac():
        """
        Argumentos:
        * a: numerador fracción 1
        * b: denominador fracción 1
        * c: numerador fracción 2
        * d: denominador fracción 2
        Te resta las dos fracciones. Te devuelve una nueva fracción simplificada por mínimo común múltiplo.
        """
        a=input('Input numerator 1:')
        b=input('Input denominator 1:')
        c=input('Input numerator 2:')
        d=input('Input denominator 2:')
        while (a.isdigit()==False) or (a=='0') or (b.isdigit()==False) or (b=='0') or (c.isdigit()==False) or (c=='0') or (d.isdigit()==False) or (d=='0'):
            a=input('Input numerator 1:')
            b=input('Input denominator 1:')
            c=input('Input numerator 2:')
            d=input('Input denominator 2:')
        else:
            a=int(a)
            b=int(b)
            c=int(c)
            d=int(d)

            x=int((a*d)-(c*b))
            y=int(b*d)
            x,y=fraccion.simplify_frac(x,y)
            return fraccion.escribir_frac(x,y)
    
    def dividir_frac():
        """
        Argumentos:
        * a: numerador fracción 1
        * b: denominador fracción 1
        * c: numerador fracción 2
        * d: denominador fracción 2
        Te divide las dos fracciones. Te devuelve una nueva fracción simplificada por mínimo común múltiplo.
        """
        a=input('Input numerator 1:')
        b=input('Input denominator 1:')
        c=input('Input numerator 2:')
        d=input('Input denominator 2:')
        while (a.isdigit()==False) or (a=='0') or (b.isdigit()==False) or (b=='0') or (c.isdigit()==False) or (c=='0') or (d.isdigit()==False) or (d=='0'):
            a=input('Input numerator 1:')
            b=input('Input denominator 1:')
            c=input('Input numerator 2:')
            d=input('Input denominator 2:')
        else:
            a=int(a)
            b=int(b)
            c=int(c)
            d=int(d)

            x=int((a*d))
            y=int(b*c)
            x,y=fraccion.simplify_frac(x,y)
            return fraccion.escribir_frac(x,y)
    
    def multiply_frac():
        """
        Argumentos:
        * a: numerador fracción 1
        * b: denominador fracción 1
        * c: numerador fracción 2
        * d: denominador fracción 2
        Te multiplica las dos fracciones. Te devuelve una nueva fracción simplificada por mínimo común múltiplo.
        """
        a=input('Input numerator 1:')
        b=input('Input denominator 1:')
        c=input('Input numerator 2:')
        d=input('Input denominator 2:')
        while (a.isdigit()==False) or (a=='0') or (b.isdigit()==False) or (b=='0') or (c.isdigit()==False) or (c=='0') or (d.isdigit()==False) or (d=='0'):
            a=input('Input numerator 1:')
            b=input('Input denominator 1:')
            c=input('Input numerator 2:')
            d=input('Input denominator 2:')
        else:
            a=int(a)
            b=int(b)
            c=int(c)
            d=int(d)

            x=int((a*c))
            y=int(b*d)
            x,y=fraccion.simplify_frac(x,y)
            return fraccion.escribir_frac(x,y)
```

* Busco outliers y los elimino

```python
Outliers = []
def outliers2 (df):
    for columnas in df:
        if np.issubdtype(df[columnas].dtype,np.number):
            Q1 = df[columnas].quantile(0.25)
            Q3 = df[columnas].quantile(0.75)
            IQR = Q3 - Q1
            outlierIzq = Q1-(1.5*IQR)
            outlierDer = Q3+(1.5*IQR)
            outliers = df[columnas][(df[columnas] < outlierIzq) | (df[columnas] > outlierDer)]
        else:
            continue
        if len(outliers)> 0:
            Outliers.append(columnas)
            print("La variable " + columnas + " tiene " + str(len(outliers)) + " valores outliers: ")
            print(outliers)
outliers2(df)
```

```python
def reemplazar_outliers_iqr(df, columnas):
    for columna in columnas:
        Q1 = df[columna].quantile(0.25)
        Q3 = df[columna].quantile(0.75)
        IQR = Q3 - Q1
        outliers = df[(df[columna] < (Q1 - 1.5 * IQR)) | (df[columna] > (Q3 + 1.5 * IQR))]
        df[columna] = np.where(df[columna] < (Q1 - 1.5 * IQR), Q1, df[columna])
        df[columna] = np.where(df[columna] > (Q3 + 1.5 * IQR), Q3, df[columna])
    return df

columnas=[col for col in df.columns if np.issubdtype(df[col].dtype,np.number)]
df=reemplazar_outliers_iqr(df, columnas)
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.head()
```

```python
from scipy.stats import shapiro
import numpy as np
for col in iris.columns:
    if np.issubdtype(iris[col].dtype,np.number):
        shapiro_test = shapiro(iris[col])
        print(f"El p-valor obtenido para {col} en el test de Shapiro-Wilk es de {shapiro_test[1]}")
```

Para la anchura del sépalo no se rechaza la null hypothesis por lo que seguiría una distribución normal.

```python
from scipy.stats import ttest_ind
data1=iris[iris['species']=='versicolor']['sepal_width']
data2=iris[iris['species']=='virginica']['sepal_width']
stat,p = ttest_ind(data1,data2)
print('stat=%.3f, p=%.3f' % (stat,p))
if p>0.05:
    print('Probablemente tengan la misma media')
else:
    print('Probablemente tengan distinta media')
```

Como sigue una distribución normal, utilizo T-test student para estuidar la media de ambas especies (versicolor y virginica) y se determina que se rechaza la null hypothesis concluyendo que ambas especies puede que tengan distinta media en su distribución para la anchura del sépalo.

```python
from scipy.stats import mannwhitneyu
data1=iris[iris['species']=='versicolor']['petal_width']
data2=iris[iris['species']=='virginica']['petal_width']
MW_test = mannwhitneyu(data1, data2)
print(f"El p-valor obtenido en el test de Mann-Whitney es de {MW_test.pvalue}")
# Probablemente tengan una distinta mediana y distribución.
```

En la anchura del pétalo se ha observado que no se sigue una distribución normal, por lo que se utiliza el test de Mann-Whitney para estudiar la mediana de ambas especies (versicolor y virginica). Se rechaza la hipótesis nula por lo que se determina que ambas especies puede que sigan una distribución distinta y tengan diferente mediana.

Se hace un estudio de la serie temporal a través de los tests de Dickey-Fuller y Kwiatkowski–Phillips–Schmidt–Shin para determinar si la serie tiene raíz unitaria, es decir, si no muestra regresión hacia la media.

```python
from statsmodels.tsa.stattools import adfuller
data = sea_ice['Extent']
stat, p, lags, obs, crit, t = adfuller(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probablemente no estacionaria')
else:
	print('Probablemente estacionaria')
```

```python
from statsmodels.tsa.stattools import kpss
data = sea_ice['Extent']
stat, p, lags, crit = kpss(data)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probablemente estacionaria')
else:
	print('Probablemente no estacionaria')
```

https://www.statsmodels.org/dev/examples/notebooks/generated/stationarity_detrending_adf_kpss.html

Se observa que KPSS indica que la serie temporal tiene raíz unitaria mientras que el test ADF indica que se trata de una serie estacionaria. ##### Búsqueda de parámetros óptimos

https://machinelearningmastery.com/grid-search-arima-hyperparameters-with-python/

```python
import warnings
from math import sqrt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

def evaluate_arima_model(X, arima_order):
    # Prepare training and test datasets
    train_size = int(len(X) * 0.66)
    train, test = X[0:train_size], X[train_size:]
    history = [x for x in train]
    
    # Make predictions
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit()
        yhat = model_fit.forecast()[0]
        predictions.append(yhat)
        history.append(test[t])
    
    # Calculate root mean squared error (RMSE)
    rmse = sqrt(mean_squared_error(test, predictions))
    return rmse

def evaluate_models(dataset, p_values, d_values, q_values):
    dataset = dataset.astype('float32')
    best_score, best_cfg = float("inf"), None
    
    for p in p_values:
        for d in d_values:
            for q in q_values:
                order = (p, d, q)
                try:
                    rmse = evaluate_arima_model(dataset, order)
                    if rmse < best_score:
                        best_score, best_cfg = rmse, order
                    print('ARIMA%s RMSE=%.3f' % (order, rmse))
                except:
                    continue
    print('Best ARIMA%s RMSE=%.3f' % (best_cfg, best_score))

# Evaluación de parámetros
p_values = range(0,3)
d_values = range(0, 3)
q_values = range(0, 3)

warnings.filterwarnings("ignore")
evaluate_models(shampoo.values, p_values, d_values, q_values)
```

Los mejores parámetros para las predicciones con ARIMA son p=1, d=2 y q=2

```python
import seaborn as sns
plt.figure(figsize=(21,6))
plt.plot(train, color = "black")
plt.plot(test, color = "red")
plt.title("Train/Test split for Shampoo Sales")
plt.ylabel("Sales Number")
plt.xlabel('Year-Month')
sns.set()
plt.show()
```

* Nueva Fecha

```python
fechas = pd.date_range(start="01-2004", end="02-2004", freq='M')
nuevas_fechas = pd.DataFrame(index=fechas)
nuevas_fechas.head()
```

```python
test = pd.concat([test, nuevas_fechas])

test
```

* Modelo

```python
from pmdarima.arima import auto_arima
model=auto_arima(train,start_p=0,d=1,start_q=0,
          max_p=5,max_d=5,max_q=5, start_P=0,
          D=1, start_Q=0, max_P=5,max_D=5,
          max_Q=5, m=12, seasonal=True,
          error_action='warn',trace=True,
          supress_warnings=True,stepwise=True,
          random_state=20,n_fits=50)
```

```python
model.fit(train)
# Creamos predicciones sobre el modelo entrenado
forecast = model.predict(n_periods=len(test))
forecast = pd.DataFrame(forecast,index = forecast.index,columns=['Prediction'])
forecast
```

```python
# Visualizar datos de entrenamiento vs datos de testeo vs el forecast
plt.figure(figsize=(21,6))
#plt.plot(train.values, color = "black")
plt.plot(test.values, color = "red")
plt.plot(forecast.values, color = "green")

plt.title("Train/Test split for Shampoo Sales")
plt.ylabel("Sales Number")
plt.xlabel('Year-Month')
sns.set()
plt.show()
```

* Errores

```python
from math import sqrt
from sklearn.metrics import mean_squared_error
test_valid = test.dropna()
forecast_valid = forecast.iloc[:-1,:]
rms = sqrt(mean_squared_error(test_valid, forecast_valid))
print("RMSE:", rms)
```

Finalmente se observa que se obtiene un error muy alto, de más del 10% del número de ventas totales, no siendo capaz de capturar de forma satisfactoria la dinámica real de la serie temporal.

* NeuralProphet

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from neuralprophet import NeuralProphet
```

```python
shampoo=pd.read_csv(r'https://raw.githubusercontent.com/jbrownlee/Datasets/master/shampoo.csv')
shampoo=shampoo.drop('Month',axis=1)
shampoo['Month']=pd.date_range(start='2001-01', end='2004-01',freq='M')
shampoo.rename(columns = {'Month':'ds', 'Sales':'y'}, inplace = True)
shampoo['ds'] = pd.DatetimeIndex(shampoo['ds'])
shampoo.head()
```

```python
shampoo.info()
```

```python
import numpy as np 
asd = np.array([0,1])
```

```python
# Instanciamos clase
m = NeuralProphet()
# Partimos datos especificando unidad mínima de tiempo
df_train, df_val = m.split_df(shampoo, freq='M', valid_p = 0.2)
# Entrenamos modelo
metrics = m.fit(df_train, freq='M', validation_df=df_val)
```

```python
metrics.plot(figsize=(15,4))
```

```python
future = m.make_future_dataframe(shampoo, periods=24, n_historic_predictions=len(shampoo))
forecast = m.predict(future)
```

```python
forecast.head()
```

```python
forecast['residual1']=None
for i in forecast.index:
    forecast.loc[i,'residual1']=forecast.loc[i,'y']-forecast.loc[i,'yhat1']
```

```python
# Creamos gráfico
f = plt.figure(figsize=(15, 8), dpi=80)
# Preparamos primer componente - Diferencia entre el actual y predicho
ax1 = f.add_subplot(221)
ax1.set_title('Residuos del modelo')
ax1.plot(forecast['residual1'])
# Preparamos segundo componente
ax2 = f.add_subplot(222)
ax2.set_title('Serie vs predicción')
ax2.plot(forecast['y'])
ax2.plot(forecast['yhat1'])
ax2.legend(['Serie original', 'Estimación'])
# Preparamos tercer componente
ax3 = f.add_subplot(223)
ax3.set_title('Tendencia de la serie')
ax3.plot(forecast['trend'])
# Preparamos cuarto componente
ax4 = f.add_subplot(224)
ax4.set_title('Temporalidad anual de la serie')
ax4.plot(forecast['season_yearly'])
```

Se observa que sigue habiendo un error considerable, pero a través del uso de NeuralProphet, conseguimos que los valores predichos se adapten un poco mejor a la serie temporal original.
